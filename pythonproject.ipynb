```python
import mysql.connector
mydb=mysql.connector.connect(
     host="localhost",
     user="root",
     password="")
```

```python
mycursor=mydb.cursor(buffered=True)
```

```python
#CREATING DATABASE
mycursor.execute("CREATE DATABASE schooldatabase")
```

```python
#SECELECTING DATABASE
mycursor.execute("USE schooldatabase")
```

```python
#CREATING TABLE-1 student details
mycursor.execute("CREATE TABLE studentdetails(ID int(3) NOT NULL UNIQUE PRIMARY KEY,NAME TEXT(50) NOT NULL,AADHARCARD VARCHAR(12) NOT NULL,PARENTSNAME TEXT NOT NULL,PHONENO VARCHAR(10) NOT NULL,CLASS VARCHAR(20) NOT NULL,SECTION VARCHAR(10) NOT NULL)")
```

```python
#creating TABLE-2  for marks
mycursor.execute("CREATE TABLE subjects(STUDENT_ID INT(3) UNIQUE,MATHSMARK INT(3),SCIENCEMARK INT(3),SOCIALMARK INT(3),LANGUAGEMARK INT(3),FOREIGN KEY (STUDENT_ID)REFERENCES studentdetails(ID))")
```

```python
#creating TABLE-3 marksheet
mycursor.execute("CREATE TABLE MARKSATEMENT(STUDENT_ID INT(3) UNIQUE,QUARTLYMARK INT(10),HALFLYMARK INT(10),ANNUALMARK INT(10),FOREIGN KEY (STUDENT_ID)REFERENCES studentdetails(ID))")
                  
```
